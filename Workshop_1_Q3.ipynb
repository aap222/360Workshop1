```python
def simulate(Δt, x, u):
    x += Δt * u
    return x
```

```python
def control(t, y):
    ### WRITE YOUR CONTROL POLICY HERE:
    ux = sin(5 * t)
    uy = cos(5 * t)
    uz = 1
    
    return array([ux, uy, uz])
```
